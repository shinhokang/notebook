```python
import bs4
import requests
```

```python
# response = requests.get( 'http://finance.naver.com/sise/dividend_list.nhn' )  
response = requests.get( 'http://finance.naver.com/sise/dividend_list.nhn?&page=2' )  

soup = bs4.BeautifulSoup(response.text)  
```

```python
stocks = []
for tr in soup.find_all('tr')[2:]:
    if ((tr.get('class') != None) and ('tr_frst' not in tr.get('class')) and ('tr_last' not in tr.get('class'))):
        tds = tr.find_all('td')
        a = tds[0].find('a')
        if (a != None):
            stocks.append({'Code':a.get('href')[-6:], 
             'Name':tds[0].text,
             '기준월':tds[2].text,
             '배당금':tds[3].text,
             '현재가':tds[1].text,
             '배당수익률':tds[4].text})
```

```python
import csv

keys = stocks[0].keys()
with open('dividend.csv', 'w') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(stocks)
```
